```python
# 01_EDA.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement des donn√©es
df = pd.read_csv('../data/tweets.csv')

# Aper√ßu g√©n√©ral
print("Dimensions :", df.shape)
print("\nColonnes :", df.columns.tolist())
print("\nTypes de donn√©es :\n", df.dtypes)
print("\nExtrait :\n", df.head())

# Valeurs manquantes
print("\nValeurs manquantes :\n", df.isnull().sum())

# Doublons
print("\nNombre de doublons :", df.duplicated().sum())

# Longueur des tweets
df['text_length'] = df['text'].astype(str).apply(len)

print("\nLongueur des tweets :")
print("Min :", df['text_length'].min())
print("Max :", df['text_length'].max())
print("Moyenne :", df['text_length'].mean())

# Distribution des longueurs
sns.histplot(df['text_length'], bins=50)
plt.title("Distribution des longueurs de tweets")
plt.xlabel("Longueur")
plt.ylabel("Nombre de tweets")
plt.show()

# R√©partition des classes
print("\nR√©partition des classes :\n", df['target'].value_counts())

sns.countplot(x='target', data=df)
plt.title("Distribution de la variable cible")
plt.xlabel("Classe (0 = normal, 1 = catastrophe)")
plt.ylabel("Nombre de tweets")
plt.show()
```
